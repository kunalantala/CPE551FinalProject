**Description:** Main notebook for the stock portfolio tracker. Imports all modules, builds a portfolio, fetches live prices, generates a report, plots a chart, saves to CSV, and streams prices using a generator.

```python
# Imports all modules needed to run the portfolio tracker
from stock import Stock
from portfolio import Portfolio
from data_handler import fetch_data, save_portfolio, load_portfolio, price_gen
from report import report_gen, portfolio_plot
```

```python
# Creates a portfolio and adds three stocks with their purchase prices
p = Portfolio("Alex & Kunal Portfolio")
p.add_stock(Stock("NVDA", 5, 500.0))
p.add_stock(Stock("AMD", 10, 120.0))
p.add_stock(Stock("TSLA", 8, 220.0))
```

```python
# Fetches live prices from Yahoo Finance and updates each stock
fetch_data(p)
```

```python
# Prints a full report of the portfolio and plots a bar chart of current values
report_gen(p)
portfolio_plot(p)
```

```python
# Saves the portfolio to a CSV file so it can be loaded later
save_portfolio(p)
```

```python
# Uses the price generator to stream each stock's symbol and price one at a time
for symbol, price in price_gen(p):
    print(f"{symbol}: ${price:.2f}")
```

```python
# Prints the portfolio summary through the defined __str__
print(p)
```

```python
# Loads the portfolio back from the CSV to verify save and load works correctly
p2 = Portfolio("Loaded Portfolio")
load_portfolio(p2)
print(p2.get_symbols()) # Prints the tuple of symbols loaded from the file
```
